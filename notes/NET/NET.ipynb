#### Part 1. Fluorescence 

* **Fluorophors** are the molecule that can absorb energy (in a specific wavelength) ans re-emits energy at different **specific** wavelength
  1. Light energy is absorted. Electron moves to a higher energy orbital 
  2. *Something* happens inside the molecule
      3. The energy is re-emitted in form of a photon of a [different wavelength](#length_difference)
* **Molecular electronic transitions** take place when electrons in a molecule are excited from **one energy level to a higher energy level**. The energy change associated with this transition provides information on the structure of a molecule and determines many molecular properties such as color.
* Not all wavelength are appropriate for fluorophors.
  * $\lambda<200 nm$ - energy $E=\frac{hc}{\lambda}$ is high enough to ionize molecule (can broke $C-C$ bonds)
  * $\lambda>700 nm$ - only low lying S1 states can be reached; typically more reactive ???
$\Rightarrow$ such fluorophors have low thermal and photochemical stability    
    **Why fluorophors with such lambda have low thermal and photochemical stability?**
  * $300 nm<\lambda<700 nm$ for absorption/emission, stable and bright fluorophors
  
 <a id='length_difference'></a> 
 
* Why the wavelength of re-emition different? 
<img  src="./images/l5g1.png"/>
The spin of the exited electron can be the same as it was when the electrone was not exited (**singlet**) or opposite (**triplet**)

 So $S_1$ - is the first singlet (+one energy level, spin the same), $S_2$ - is the second singlet(+two energy levels, spin the same) 
 
 $T_1$ is the first triplet (+one energy level, reversed spin)
  
  * **Internal conversion** is a transition from a higher to a lower electronic state in a molecule or aom. The change of energy is due to **vibrational relaxation** (The excitation energy is transformed into heat)
  
  * **Intersystem crossing ** is the excitation to $T_1$, so it includes changing of electron's spin. In from the new state it is hard to get back to $S_0$, because we will get  "forbidden" energy state (two electrons with the same spin) $\Rightarrow$ the lifetime of electrone in $T_1$ state is $100 \ s$, whereas $S_1$ is only $10^{-7}-10^{-9} \ s$
* Important parameters of fluorescent dyes
  * $\varepsilon$ - molar excitation coefficient is a measurement of how strongly a molecules **absorb light** at a given wavelength.
    
    Beer-Lambert law: $A=\varepsilon cL$, where $A$ - absorbance, $c$ - concentation, $L$ - pathlength
    <img  src="./images/l5f2.jpg"/>
    The beam radiant power becomes weaker as it passes through solution
  * Quantum yield : $ \Phi_{f}$ $ =\frac{N_{emitted}}{N_{absorbed}}$
  * Fluorescence lifetime $\tau_f$ - average **time** a molecule spends in its excited
(singlet) state **before emission**
occurs

      - first uncharged $\rightarrow$ can get to the cell, when lipophilic blocking groups are cut by

   3. via bulk or single-cell electroporation
   4. by application of crystals of lipophilic dyes to the cell membrane (e.g. DiO, DiI)

    3. chemical [permeabilization][2]

#### Part 4. Super-Resolution Microscopy.

**RESOLFT = REversible Saturable Optical Fluorescence Transition**
- group of techniques based on the fact that emission of a fluorophor typically has a non-linear
dependence on:
 1. excitation energy
 2. wavelength
 3. spatial distribution of excitation light
 
- use ability to switch off and on with fluorescence (via one of the dependence above)

**STED**:

- resolution increase by:
de-excitation of excited fluorophors via
stimulated emission using a red-shifted
light beam with a local intensity zero
in the center (“donut mode beam”)
<img  src="./images/l6f5.png"/>

- Advantages:
  - pulsed laser not necessary
  - is compatible with photostable FP variants **???**  $\Rightarrow$ live cell imaging
- Disadvantage
  - requires relatively high intensities for
efficient stimulated emission (**photodamage**)

- sharpen or decrease size of PSF 




Also make use of the fact that single fluorophors can actually be located
with **100% precision** if **sparse enough** (and S/N ration excellent, etc.)


**Single Marker Switching (SMS) Microscopy**

- Image acquisition process:
all dyes off $\rightarrow$ switch-on small fraction
$\rightarrow$ acquire image $\rightarrow$ all off 

- Advanteges: 
  - **~20 nm** in the x-y plane
  - **wide-field** technique
- Disadvanteges: 
  - not (yet) generally applicable to living
tissue; used **photo-switchable dyes**
often need **specific conditions**
  - very bad temporal resolution, since
many (up to 1000) frames need to
be acquired per reconstructed image

Opsins


> Opsin genes are divided into two distinct superfamilies: **microbial opsins (type I)** and **animal

- Sequence homology between the two opsin families is very low (So they are quite different)

> - type II opsins(e.g. rhodopsin):
  - (in higher eukaryotes, mainly involved in vision)
  - G protein-coupled receptors, binding 11-cis retinal (in darkness), which isomerizes to all-trans
retinal upon illumination and initiates intra-cellular signaling (i.e. phototransduction). Finally,
all-trans retinal is replace by 11-cis.

#### Part 1. Strength-duration relation

- We want inject current, that is enough to depolarize cell membrane above threshold $\Delta V_{threshold}$ 
- Using $RC$-circuit: $$\frac{\Delta V_{threshold}}{R_M}=I_{rheobase}=I_{threshold}\left(1-e^{\frac{t}{\tau}}\right)$$
 <img  src="./images/l9g1.PNG" width=400 />


1. Assumption that the cell **membrane is linear** up to action potential threshold
2. Intracellular stimulation of spherical cell. In most applications,
stimulating electrodes are extracellular and produce a response which depends
on electrode location as well as on cell geometry.
3. Fixed threshold does not account for short stimuli (duration << activation time
for sodium ion channels)
4. Fixed threshold does not account for long stimuli (channel inactivation time constant)

So we use this theory:
 - to explain why there is no AP, even if stimulus is long (because $I<I_{rheobase}$)
 - to explain that we need **more strong** stimulus if the duration is short. 

Electrode considerations
- We can inject electrons (faradaic currents) 
  - This creates a lot of problems, because electrons involved in many (irreversible) reactions. So we don't know exact amount of injected current 
- We can charge capacitance to induce ion current in the direction of capacitance (capacitive charge injection)
