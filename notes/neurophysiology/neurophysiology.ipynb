#### Part 2. GHK equation
*	Assuming **constant membrane potential V** and **independence of different ion types** one can derive current density  or Goldman-Hodgkin-Katz (GHK) current equation:

       $$j_A=P_Az_AF\frac{z_AFV}{RT}\left ( \frac{[A]_{in}-[A]_{out}\cdot \exp{\frac{-z_A FV}{RT}}}{1- \exp{\frac{-z_A FV}{RT}}}\right) $$
        
        
*	Setting **net current to 0**, compute the equilibrium potential Goldman-Hodgkin-Katz (GHK) voltage equation:

        $$E_m=\frac{RT}{F}\ln \left (\frac{P_{K^+}[K^+]_{out}+P_{Na^+}[Na^+]_{out}+P_{Cl^-}[Cl^-]_{in}}{P_{K^+}[K^+]_{in}+P_{Na^+}[Na^+]_{in}+P_{Cl^-}[Cl^-]_{out}}\right)$$
